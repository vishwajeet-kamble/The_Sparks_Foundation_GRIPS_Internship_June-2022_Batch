### Sub-Categories

```python
## Sub-Categories

Sub_Categories = df['Sub-Category'].unique()
print(Sub_Categories, '\n')

print('Sub-Categories are :' ,'\n')
for i in Sub_Categories:
    print(i)

print()
print('In total there are ', len(Sub_Categories) ,' Sub-Categories')

## Plotting Countplot for Sub-Categories 

sns.set_style('darkgrid')
plt.figure(figsize = (18,5))
sns.countplot(x = df['Sub-Category'], order = df['Sub-Category'].value_counts().index, edgecolor = 'k')
plt.xlabel('Sub-Category', fontsize = 18)
plt.ylabel('Sales', fontsize = 18)
plt.title('Sub-Category wise Sales', fontsize=25)
plt.xticks(rotation=90, fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()
```

### Region wise Sales Overrview

```python
sales_region = df[['Region','Sales']]


sales_region.groupby("Region").sum().sort_values(by = 'Sales', ascending=False).plot(kind="bar", figsize=(15,5), color = 'Orange' , edgecolor = 'k')
plt.title('Region wise Sales Overrview', fontsize = 20)
plt.xlabel('Regions', fontsize = 18)
plt.ylabel('Sales', fontsize = 18)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()

print('From Visualisation we can say :','\n')
print('West Region have High Compare to Other Regions')
```

### Sales Overview By State

```python
## Sales By State

sales_states = df[['State','Sales']]

sales_states.groupby("State").sum().sort_values(by = 'Sales', ascending=False).plot(kind="bar", figsize=(25,8), color='Purple')

plt.title('State wise Sales Overrview', fontsize = 23)
plt.xlabel('State', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()

print('From Visualisation we can say :','\n')
print('Sales in California State is High Compare to Other States')
```

### State wise Profit and Loss

```python
## State wise Profit and Loss

plt.figure(figsize=(25,8))
plt.xlabel('State', fontsize = 20)
plt.ylabel('Profit', fontsize = 20)
plt.xticks(rotation=90, fontsize = 17)
plt.yticks(fontsize = 12)
plt.title('State wise Profit and Loss', fontsize = 23)
sns.set_style('whitegrid') 
sns.barplot(x=df['State'], y=df['Profit'])

print('State Vermount has Highest Profit')
```

### Sales of Top 10 Cities

```python
## Sales of Top 10 Cities

top10 = df.groupby('City')['Sales'].sum().sort_values(ascending = False).iloc[0:10]

plt.figure(figsize=(15,6))
plt.title('Sales of Top 10 Cities', fontsize = 20)
plt.xlabel('City', fontsize = 20)
plt.ylabel('Total sum of Sales', fontsize = 18)
plt.xticks(rotation=90, fontsize = 16)
plt.yticks(fontsize = 15)
top10.plot(kind='bar', color = 'Violet', edgecolor = 'k')


print('We can see New York City has Highest Sales Value')
```

### Statewise Sales and Profit overview

```python
## Statewise Sales and Profit overview

overview = df.groupby('State')['Profit','Sales'].sum()
overview.plot(kind='bar',figsize=(27,8))
plt.xlabel('State', fontsize = 20)
plt.ylabel('Profit and Sales', fontsize = 20)
plt.xticks(rotation=90, fontsize = 18)
plt.yticks(fontsize = 15)
plt.title('Sales and Profit overview', fontsize = 23)
plt.show()
```

```python
pd.crosstab(df['Sub-Category'], df.Region).plot(kind='area',figsize=(25,8))
plt.title('Sub-Category Sales Regionwise', fontsize = 25)
plt.xlabel('Sub-Category', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.xticks(rotation=90, fontsize = 20)
plt.yticks(fontsize = 15)
plt.legend(borderpad=2, fontsize=20)
plt.show()
```

```python
pd.crosstab(df['Category'], df.Region).plot(kind='area',figsize=(25,8))
plt.title('Sub-Category Sales Regionwise', fontsize = 25)
plt.xlabel('Category', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.xticks(rotation=90, fontsize = 20)
plt.yticks(fontsize = 15)
plt.legend(borderpad=2, fontsize=20)
plt.show()
```

```python
pd.crosstab(df['Segment'], df.Region).plot(kind='area',figsize=(25,8))
plt.title('Segmentwise Sales Regionwise', fontsize = 25)
plt.xlabel('Segment', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.xticks(rotation=90, fontsize = 20)
plt.yticks(fontsize = 15)
plt.legend(borderpad=2, fontsize=20)
plt.show()
```

```python
pd.crosstab(df['Discount'], df.Region).plot(kind='area',figsize=(25,8))
plt.title('Sales for each Discount Regionwise', fontsize = 25)
plt.xlabel('Discount', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.xticks(rotation=90, fontsize = 20)
plt.yticks(fontsize = 15)
plt.legend(borderpad=2, fontsize=20)
plt.show()
```
